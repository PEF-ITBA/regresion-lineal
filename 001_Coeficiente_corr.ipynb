```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
N = 10000
np.random.seed(42)
df['X1'] = np.random.normal(0, 1, N)
df['X2'] = df['X1']  + np.random.normal(0, 1, N)
df['X3'] = df['X1']**2
df['X4'] = df['X1']**3
df['X5'] = np.abs(df['X1'])
df['X6'] = df['X1']**2 + np.random.normal(0, 0.5, N)
```

```python
df.corr()
```

```python
from pandas.plotting import scatter_matrix
```
